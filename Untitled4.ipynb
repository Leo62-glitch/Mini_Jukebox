```python
!pip install transformers

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer = AutoTokenizer.from_pretrained("gpt2")  # Utilise gpt2 ou un modèle plus léger pour tester
model = AutoModelForCausalLM.from_pretrained("gpt2", device_map="auto")  # Modèle léger pour éviter la surcharge

# Fonction pour générer la réponse en streaming
def generate_response_streaming(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors='pt').to('cuda' if torch.cuda.is_available() else 'cpu')
    model.to('cuda' if torch.cuda.is_available() else 'cpu')

    # Génération par étapes avec "streaming"
    output_stream = model.generate(input_ids, max_length=100, do_sample=True, pad_token_id=tokenizer.eos_token_id)

    # Décoder le flux de tokens un par un pour les afficher au fur et à mesure
    for token in output_stream[0]:
        decoded_token = tokenizer.decode(token.unsqueeze(0))
        print(decoded_token, end='', flush=True)  # Affiche progressivement sans nouvelle ligne

if __name__ == "__main__":
    user_input = input("You: ")
    print("Assistant:", end='', flush=True)
    generate_response_streaming(user_input)
    print("\n")  # Pour ajouter une ligne à la fin de la génération
```
